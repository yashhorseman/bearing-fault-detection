```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter
from scipy.fft import fft, fftfreq
```

```python
fs = 12000  
t = np.linspace(0, 0.5, fs//2, endpoint=False)
sig = np.sin(2*np.pi*50*t) + 0.4*np.random.randn(len(t))
```

```python
b, a = butter(4, 200, btype='low', fs=fs)
filt = lfilter(b, a, sig)
```

```python
f = fftfreq(len(t), 1/fs)[:len(t)//2]
mag = np.abs(fft(filt))[:len(t)//2]
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(t[:500], sig[:500], alpha=0.5, label='raw')
plt.plot(t[:500], filt[:500], 'r', label='filt')
plt.legend(); plt.title('Time')
plt.subplot(1, 2, 2)
plt.plot(f[:300], mag[:300])
plt.title('Freq (50Hz peak)')
plt.show()
```
